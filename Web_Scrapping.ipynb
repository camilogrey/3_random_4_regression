# Introducción al web scraping

## ¿Qué es el web scraping?

El web scraping es el proceso de extracción automatizada de datos de páginas web. En términos simples, implica escribir un programa o script para recorrer una o varias páginas web, analizar su contenido y extraer la información deseada. Esta información puede ser texto, imágenes, enlaces, datos tabulares u otros tipos de contenido.

El web scraping es utilizado en una variedad de aplicaciones, como recopilación de datos para análisis, monitoreo de precios en línea, agregación de contenido, seguimiento de cambios en sitios web y mucho más. Permite a los usuarios obtener grandes cantidades de datos de manera eficiente y automatizada, en lugar de hacerlo manualmente, lo que sería muy laborioso y propenso a errores.

Sin embargo, es importante destacar que el web scraping debe realizarse de manera ética y respetando los términos de servicio de los sitios web objetivo. Algunos sitios web pueden tener políticas específicas sobre el acceso automatizado a sus datos, y el scraping sin permiso puede considerarse una violación de sus términos de uso. Además, en algunos casos, el web scraping puede tener implicaciones legales, por lo que es importante estar al tanto de las regulaciones locales y consultar con un asesor legal si es necesario.

## ¿Por qué utilizar el web scraping?

1. Acceso a información no estructurada: Internet alberga una gran cantidad de información valiosa que no se encuentra disponible en bases de datos estructuradas o APIs. El web scraping permite acceder a estos datos sin procesar y transformarlos en formatos útiles para análisis, investigación o toma de decisiones.

2. Monitoreo de precios y tendencias: Empresas y organizaciones pueden utilizar el web scraping para monitorizar precios de productos, tendencias del mercado o cambios en la competencia. Esto les permite ajustar sus estrategias de negocio de manera proactiva y tomar decisiones informadas.

3. Investigación y análisis de datos: Investigadores y analistas pueden recopilar datos de diversas fuentes web para realizar estudios de mercado, análisis de opinión pública o investigaciones académicas. El web scraping les permite obtener información de una manera eficiente y escalable.

4. Automatización de tareas: El web scraping puede automatizar tareas repetitivas como la recopilación de noticias, la actualización de bases de datos o la extracción de información de formularios web. Esto libera tiempo para que los empleados se enfoquen en tareas más estratégicas.

5. Creación de conjuntos de datos personalizados: El web scraping permite crear conjuntos de datos personalizados para entrenar modelos de aprendizaje automático, realizar análisis de big data o desarrollar aplicaciones web.

6. Enriquecimiento de datos existentes: El web scraping se puede utilizar para complementar y enriquecer datos existentes en bases de datos o sistemas CRM. Esto permite obtener una visión más completa de clientes, productos o mercados.

**Herramientas y técnicas básicas de web scraping en Python**

Para realizar web scraping en Python, existen diversas herramientas y técnicas que facilitan el proceso de extracción y análisis de datos. A continuación, se presenta una descripción general de las herramientas y técnicas básicas:

**Herramientas:**

* Requests: Esta biblioteca es fundamental para realizar solicitudes HTTP a páginas web y obtener el contenido HTML. Permite enviar diferentes tipos de solicitudes, establecer encabezados personalizados y manejar redirecciones.

* Beautiful Soup: Esta biblioteca facilita el análisis y parseo del contenido HTML obtenido de las páginas web. Permite navegar por la estructura del árbol HTML, encontrar elementos específicos y extraer datos relevantes.

* Lxml: Es una biblioteca de análisis XML que puede ser útil para extraer datos de sitios web que utilizan este formato. Ofrece funciones eficientes para procesar documentos XML y extraer información relevante.

* Regular Expressions: Las expresiones regulares pueden ser útiles para extraer datos específicos que no se ajustan a una estructura HTML estándar. Permiten buscar patrones en cadenas de texto y extraer información relevante.

**Técnicas:**

* Análisis de HTML estático: Este enfoque implica analizar directamente el código HTML de la página web para identificar y extraer los datos deseados. Se utiliza Beautiful Soup para navegar por la estructura HTML y encontrar los elementos que contienen la información relevante.

* Análisis de JavaScript: Algunas páginas web utilizan JavaScript para cargar contenido dinámicamente o para ocultar datos. Para extraer estos datos, es necesario utilizar herramientas como Selenium o Puppeteer que permiten simular la interacción de un usuario en el navegador y capturar el contenido dinámico.

* APIs: Algunos sitios web ofrecen APIs que permiten acceder a sus datos de forma programada. Si un sitio web proporciona una API, se puede utilizar en lugar de realizar scraping, ya que es un método más oficial y controlado para obtener datos.



# Fundamentos de HTML y CSS

###  HTML (HyperText Markup Language):

HTML es el lenguaje de marcado estándar utilizado para crear páginas web. Proporciona la estructura y el contenido de una página web, utilizando una serie de elementos y etiquetas predefinidas que definen cómo se visualiza y se organiza la información en el navegador.

**Elementos:** Los elementos son los componentes básicos de HTML y están rodeados por etiquetas. Cada elemento tiene una etiqueta de apertura y una etiqueta de cierre, y puede contener texto y otros elementos.

<`p>Este es un párrafo de ejemplo.</p`>

**Atributos:** Los atributos proporcionan información adicional sobre un elemento y se especifican dentro de la etiqueta de apertura. Los atributos pueden incluir cosas como identificadores únicos (id), clases de estilo (class), enlaces (href), entre otros.

`<a href="https://www.ejemplo.com">Enlace a Ejemplo</a>`

**Estructura básica:** Toda página web HTML comienza con una estructura básica que incluye elementos como `<html>`, `<head>` y `<body>`. La mayoría de los elementos se colocan dentro del elemento `<body>`, que contiene el contenido visible de la página.
```
<!DOCTYPE html>
<html>
  <head>
    <title>Título de la página</title>
  </head>
  <body>
    <h1>Encabezado Principal</h1>
    <p>Este es un párrafo de ejemplo.</p>
  </body>
</html>
```



* `<header>`: la etiqueta de encabezamiento define el contenido introductorio de una página o sección.
* `<nav>`: la etiqueta de navegación se utiliza para los enlaces de navegación. Puede anidarse dentro de la etiqueta `<header>`, pero las etiquetas `<nav>` de navegación secundaria también suelen usarse en otras secciones de la página.
* `<main>`: esta etiqueta contiene el contenido principal (también llamado cuerpo) de una página. Solo debería haber una etiqueta `<main>` por página.
* `<article>`: la etiqueta artículo define el contenido independiente de la página o web en la que se encuentra. No significa necesariamente una "entrada del blog". Piensa en ella más bien como "una prenda de vestir": una etiqueta independiente que puede emplearse en varios contextos.
* `<section>`: utilizar `<section>` es una forma de agrupar contenidos que tratan sobre temas similares. Las etiquetas de sección son diferentes a las de artículo, no son necesariamente autónomas, sino que forman parte de algo más.
* `<aside>`: un elemento "aside" define el contenido de menor importancia. Suele emplearse para las barras laterales, que incluyen información complementaria pero no esencial.
* `<footer>`: usa `<footer>` en la parte inferior de una página. La etiqueta suele incluir información de contacto, derechos de autor y navegación por la web.

### Etiquetas HTML Semánticas para Texto

* `<h1>` (encabezado): la etiqueta H1 marca el encabezado de nivel superior. Normalmente, solo hay un título H1 por página.
* `<h2> a <h6>` (subtítulos): los subtítulos con varios niveles de importancia. Puede haber varios subtítulos del mismo nivel en una misma página.
* `<p>` (párrafo): un párrafo de texto independiente.
* `<a>` (anchor): se utiliza para marcar un hiperenlace de una página a otra.
* `<ol>` (lista ordenada): lista de elementos que se muestran en un orden determinado, empezando por las viñetas. Las etiquetas
* `<li>` (elemento de lista) contienen un único elemento de la lista.
* `<ul>` (lista desordenada): lista de elementos que no necesitan mostrarse en un orden determinado, empezando por los números ordinales. Las etiquetas `<li>` (elemento de lista) contienen un único elemento de la lista.
* `<q> / <blockquote>`: una cita del texto. Utiliza `<blockquote>` para citas largas, de varias líneas, y `<q>` para citas más cortas, en línea.

### CSS (Cascading Style Sheets):

CSS es un lenguaje de estilo utilizado para controlar el aspecto y la presentación de una página web. Permite definir el diseño, los colores, las fuentes y otros aspectos visuales de los elementos HTML, separando así la estructura del contenido de su presentación.

* **Reglas de estilo**: Las reglas de estilo en CSS consisten en un selector y un bloque de declaración que define cómo se deben mostrar los elementos seleccionados. Los selectores pueden ser etiquetas HTML, clases, identificadores u otros criterios.


```
h1 {
  color: blue;
  font-size: 24px;
}
```

* **Clases y ID**: Las clases y los identificadores son atributos que se pueden aplicar a elementos HTML para aplicar estilos específicos a esos elementos. Las clases se pueden reutilizar en múltiples elementos, mientras que los identificadores deben ser únicos en toda la página.

**html**
```
<p class="destacado">Este párrafo tiene un estilo destacado.</p>
```

**css**
```
.destacado {
  font-weight: bold;
  color: red;
}
```




# Librerias importantes para web scrapping en python

1. **[Beautiful Soup](https://pypi.org/project/beautifulsoup4/)**: Es una biblioteca Python que se utiliza para extraer datos de archivos HTML y XML. Facilita la navegación por el árbol de análisis y la búsqueda de elementos utilizando selectores CSS y expresiones XPath.
2. **[Requests](https://requests.readthedocs.io/en/latest/user/quickstart/)**: Aunque no es específicamente una biblioteca de web scraping, Requests es una poderosa biblioteca para realizar solicitudes HTTP en Python. Se utiliza comúnmente junto con Beautiful Soup para descargar el contenido HTML de una página web antes de realizar el análisis de scraping.
3. [**Scrapy**](https://scrapy.org/): Es un framework de scraping de páginas web de alto nivel y orientado a objetos para Python. Proporciona un conjunto completo de herramientas para extraer datos de manera eficiente y escalable, manejar la navegación y seguir enlaces automáticamente.
4. [**Selenium**](https://selenium-python.readthedocs.io/): A diferencia de las bibliotecas anteriores, Selenium es una herramienta de automatización del navegador que permite controlar un navegador web real a través de una interfaz programática. Es útil para el scraping de páginas web que utilizan JavaScript para la representación dinámica del contenido.

5. [**Lxml**](https://lxml.de/): Es una biblioteca de Python que permite el análisis y manipulación eficiente de documentos XML y HTML. Ofrece un rendimiento superior a Beautiful Soup en algunas situaciones y es compatible con XPath.
6. **[PyQuery](https://pypi.org/project/pyquery/)**: Inspirado en jQuery, PyQuery es una biblioteca que proporciona una interfaz similar para analizar y manipular documentos HTML. Es útil si estás familiarizado con la sintaxis de jQuery y quieres aplicarla en el scraping de datos.

**Creación de un objeto BeautifulSoup:**

Para analizar un documento HTML, primero necesitas crear un objeto BeautifulSoup pasando el contenido HTML como argumento junto con el analizador que deseas utilizar (por lo general, 'html.parser'):

````python
html = "<html><head><title>Ejemplo</title></head><body><p>Hola, mundo!</p></body></html>"
soup = BeautifulSoup(html, 'html.parser')
soup
````

**Acceso a elementos HTML:**

Una vez que tienes un objeto BeautifulSoup, puedes acceder a los elementos HTML utilizando métodos como find() y find_all(). Estos métodos te permiten buscar elementos por etiqueta, clase, identificador, atributos, etc.

````python
title_tag = soup.find('title')
print(title_tag.text)  # Imprime el contenido de la etiqueta <title>

p_tags = soup.find_all('p')
for p_tag in p_tags:
    print(p_tag.text)  # Imprime el contenido de todas las etiquetas <p>
````

**Navegación por el árbol DOM:**

BeautifulSoup proporciona varios métodos para navegar por el árbol DOM de un documento HTML. Algunos de los métodos más comunes incluyen:
* .parent: Accede al elemento padre.
* .children: Itera sobre los elementos secundarios directos.
* .descendants: Itera sobre todos los elementos descendientes.
* .next_sibling y .previous_sibling: Acceden al siguiente y al anterior hermano del elemento, respectivamente.

````python
import requests


# URL de la página web
url = 'https://es.wikipedia.org/wiki/Demograf%C3%ADa_de_Colombia'
````

````python
# Realizar la solicitud HTTP y obtener el contenido de la página
response = requests.get(url)
html = response.text
print(html)
````

````python
# Crear un objeto BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
````

````python
# Encontrar la tabla de la infobox de demografía
tabla_demografia = soup.find('table', class_='infobox')
````

````python
# Encontrar todas las filas de la tabla
filas = tabla_demografia.find_all('tr')
````

El elemento básico de definición de tabla es `<TABLE> </TABLE>` y en su interior se disponen los sub elementos `<TR>` para definir una fila (Row) `<TH>` para definir una cabecera (Header) `<TD>` para definir una celda de datos (Data).

````python
# Iterar sobre las filas y extraer los datos
for fila in filas:
    # Encontrar las celdas de cada fila
    celdas = fila.find_all(['th', 'td'])
    for celda in celdas:
        # Extraer el texto de la celda y limpiarlo
        texto = celda.get_text(strip=True)
        print(texto)
    print('---------------------')
````

````python
import pandas as pd
# Extraer los datos de la tabla y almacenarlos en una lista de listas
datos_tabla = []
filas = tabla_demografia.find_all('tr')
for fila in filas:
    celdas = fila.find_all(['th', 'td'])
    fila_datos = []
    for celda in celdas:
        texto = celda.get_text(strip=True)
        fila_datos.append(texto)
    if fila_datos:
        datos_tabla.append(fila_datos)

# Crear un DataFrame con los datos extraídos
df = pd.DataFrame(datos_tabla, columns=['Indicador', 'Valor'])

# Imprimir el DataFrame
df
````

# Consideraciones éticas y legales del web scraping

El web scraping, si bien es una herramienta poderosa para la recolección de datos, debe utilizarse de manera responsable y ética, considerando las implicaciones legales y morales que conlleva. A continuación, se detallan algunos aspectos importantes a tener en cuenta:

* **Aspectos éticos:**

1. Respeto a la propiedad intelectual: El contenido de un sitio web, como texto, imágenes o videos, puede estar protegido por derechos de autor. Extraer este contenido sin autorización previa del propietario puede ser una violación a sus derechos.
2. Privacidad de datos: El web scraping puede implicar la extracción de datos personales de usuarios, como nombres, direcciones de correo electrónico o información financiera. Es crucial manejar estos datos con cuidado y cumplir con las leyes de protección de datos del pais.
3. Transparencia: Es importante ser transparente con los usuarios sobre el uso del web scraping. Informarles en la política de privacidad o en un aviso claro cómo se recopilan y utilizan sus datos.
4. Evitar dañar a los sitios web: El scraping excesivo puede sobrecargar los servidores de un sitio web y afectar su rendimiento. Es necesario realizar solicitudes de manera responsable y espaciada para evitar causar daños.

* **Aspectos legales:**

1. Términos de servicio: Los sitios web suelen tener términos de servicio que establecen las reglas para su uso. Algunos sitios web pueden prohibir explícitamente el scraping o tener limitaciones sobre la cantidad de datos que se pueden extraer. Es fundamental leer y comprender estos términos antes de realizar cualquier tipo de scraping.
2. Leyes de derechos de autor: La extracción de contenido protegido por derechos de autor sin autorización previa puede ser una violación a las leyes de derechos de autor.
3. Leyes de protección de datos: Como se mencionó anteriormente, el manejo de datos personales extraídos a través de scraping debe cumplir con las leyes de protección de datos vigentes en cada región.
4. Leyes contra el fraude y la competencia desleal: El uso del web scraping para obtener información confidencial o para dañar a la competencia puede ser ilegal.

**Recomendaciones para un web scraping ético y legal:**

1. Solicitar permiso: Si es posible, contactar a los propietarios de los sitios web y solicitar permiso para realizar scraping.
2. Limitar la tasa de solicitudes: Realizar solicitudes a un ritmo moderado para evitar sobrecargar el servidor del sitio web.
3. Identificar como un scraper: Es recomendable incluir información en las solicitudes de scraping que identifique al scraper y el propósito del scraping.
4. Manejar los datos de manera responsable: Implementar medidas de seguridad para proteger los datos extraídos y evitar su uso indebido.
5. Estar actualizado sobre las leyes y regulaciones: Mantenerse informado sobre las leyes y regulaciones relacionadas con el web scraping en las regiones donde se opera.
